```python
import pandas as pd

train_with_news = pd.read_parquet("../data/processed/train_with_news.parquet")
```

```python
train_users = sorted(train_with_news["user_id"].unique())
train_items = sorted(train_with_news["item_id"].unique())

user_idx_map = {u: i for i, u in enumerate(train_users)}
item_idx_map = {it: i for i, it in enumerate(train_items)}
idx_user_map = {i: u for u, i in user_idx_map.items()}
idx_item_map = {i: it for it, i in item_idx_map.items()}
```

```python
import json

with open("../data/processed/user_idx_map.json", "w") as f:
    json.dump(user_idx_map, f)
with open("../data/processed/item_idx_map.json", "w") as f:
    json.dump(item_idx_map, f)
with open("../data/processed/idx_user_map.json", "w") as f:
    json.dump(idx_user_map, f)
with open("../data/processed/idx_item_map.json", "w") as f:
    json.dump(idx_item_map, f)
```

```python
print("Unique train users:", len(user_idx_map))
print("Unique train items:", len(item_idx_map))
print("Matrix index space:", (len(user_idx_map), len(item_idx_map)))
```
