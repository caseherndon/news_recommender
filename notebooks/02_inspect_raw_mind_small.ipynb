```python
dev_behaviors = pd.read_csv(
    "../data/raw/MINDsmall_dev/behaviors.tsv",
    sep="\t",
    header=None,
    names=["impression_id", "user_id", "time", "history", "impressions"]
)

train_users = set(behaviors["user_id"])
dev_users = set(dev_behaviors["user_id"])
print("Dev unique users:", dev_behaviors["user_id"].nunique())
print("Users only in dev:", len(dev_users - train_users))
```

```python
def count_clicks(behaviors_df):
    clicked = 0
    not_clicked = 0
    for impressions in behaviors_df["impressions"]:
        for item in impressions.split():
            _, label = item.rsplit("-", 1)
            if label == "1":
                clicked += 1
            else:
                not_clicked += 1
    return clicked, not_clicked

train_clicked, train_not_clicked = count_clicks(behaviors)
dev_clicked, dev_not_clicked = count_clicks(dev_behaviors)

print("Train — clicked:", train_clicked, "| not clicked:", train_not_clicked)
print("Dev — clicked:", dev_clicked, "| not clicked:", dev_not_clicked)
```
